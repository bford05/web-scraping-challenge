```python
#import packages for scraping twitter
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
```

```python
#print results from twitter webpage for review
#print(soup.prettify())
```

```python
#Define parameters to query tweets to find weather conditions
begin_date = dt.date(2020, 5, 1)
end_date = dt.date(2020, 5, 3)
limit = 200
lang = 'english'
tweets = query_tweets("InSight sol", begindate = begin_date, enddate = end_date, limit=limit, lang=lang)
tweets
```

```python
#Place tweets into dataframe
tweets_df = pd.DataFrame(t.__dict__ for t in tweets)
tweets_df
```

```python
#Retrieve mars weather info from dataframe
mars_weather = tweets_df.iloc[2]['text'][:164]
mars_weather
```
